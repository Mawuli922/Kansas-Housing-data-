Determining the appropriate regression function for the prediction of prices based on Living sq. ft in Kansas 

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
lr = LinearRegression()
```

```python
y_data = df_1[["price"]]
x_data = df_1[["sqft_living"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 0)
```

```python
test_stat = []
degree_poly = [1,2,3,4]
for n in degree_poly:
    pr = PolynomialFeatures(degree=n)
    x_train1 = pr.fit_transform(x_train)
    x_test1 = pr.fit_transform(x_test)
    lr.fit(x_train1, y_train)
    test_stat.append(lr.score(x_test1,y_test))
    
    
    
    
    
```

```python
test_stat
```

from the list of R squared scores, degree 4 polynomial seems to have the highest score however the discrepancy between 2,3,4 is very minute and as such a compromise to avoid overfitting leads to selection of degree 3 polynomial for our linear model

```python
pr = PolynomialFeatures(degree = 3)
```

```python
x_train3 = pr.fit_transform(x_train)
x_test3 = pr.fit_transform(x_test)
lr.fit(x_train3,y_train)
yhat_test = lr.predict(x_test3)
yhat_train = lr.predict(x_train3)
```

```python
plt.figure(figsize = (12,8))
plt.title("Comparison of Density Distribution of Training and Test predictions against the Actual data")
sns.distplot(y_data, hist = False, color = 'r', label = "Actual Data")
sns.distplot(yhat_test, hist = False, color = 'b', label = "Test predictions")
sns.distplot(yhat_train, hist = False, color = 'g', label = "Training predictions")
plt.legend()
plt.xlabel("Kansas Housing Prices (US$)")
```

The presence of outliers within the dataset causes the distribution to be skewed left to reflect the true mean.
